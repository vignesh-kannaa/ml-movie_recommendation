## Types

There are several types of recommendation systems to suggest movies and TV shows to its users. Here are some of the main types:

**Popularity based**

It gives generalized movie recommendations to all users based on popularity and genre. It assumes that similar users will enjoy similar movies. Popular and critically acclaimed movies are more likely to be recommended as they have a higher chance of being liked by the average audience.

**Content-based filtering**

This technique recommends movies or TV shows to users based on the similarity between the attributes of items and the user's interests. For example, if a user frequently watches horror movies, the system will recommend other horror movies.

**Collaborative filtering** 

This technique recommends movies or TV shows to users based on their viewing history and ratings, as well as those of other users with similar tastes. This approach relies on user behavior and does not require any explicit information about the items themselves.


**Personalized genre ranking**

This technique recommends movies and TV shows based on a user's most preferred genres. The system will rank the genres based on the user's viewing history and present recommendations accordingly.

**Hybrid recommendation systems**

This combines both collaborative and content-based filtering techniques to provide more accurate recommendations. Hybrid systems take advantage of the strengths of both approaches and compensate for their weaknesses.

**Context-based filtering**

This technique considers external factors, such as time of day, day of the week, location, and weather, to provide recommendations based on the user's current context.

To provide more personalized movie recommendations, we can use content-based filtering. separating the required data for the necessary columns

```python
from ast import literal_eval

# converting to list of dictionaires
features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    df[feature] = df[feature].apply(literal_eval)
```

```python
get_names = lambda lst: [d["name"] for d in lst] if lst else None
df["genres"] = df["genres"].apply(get_names)
df['cast']=df["cast"].apply(get_names)
df['keywords']=df["keywords"].apply(get_names)
```

```python
get_director = lambda lst: [d['name'] for d in lst if d['job'] == 'Director'] if lst else np.nan
df['director'] = df['crew'].apply(get_director)
```

```python
df[['cast', 'director', 'keywords', 'genres']].head()
```

To prepare the data for our vectorizer, we need to standardize the format of the names and keywords. This involves converting all characters to lowercase and removing spaces between words. This step is important to ensure that the vectorizer can differentiate between different names and keywords that have similar spellings but different meanings. For example, without removing spaces, the vectorizer may count  "Johnny Depp" and "Johnny Galecki" as the same keyword because of the word 'Johnny', which would result in inaccurate vector representations.

```python
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    elif isinstance(x, str):
        return str.lower(x.replace(" ", ""))
    else:
        return ''
```

```python
features = ['cast', 'keywords', 'director', 'genres']

for feature in features:
    df[feature] = df[feature].apply(clean_data)
    
df[features]
```

To create a new feature for the vectorizer, we can combine the existing features like the movie's cast, director, genres, and keywords into a single string. This combined string can be used as an input for the vectorizer to create the count matrix. 

```python
def create_input(row):
    keywords = ' '.join(row['keywords'])
    cast = ' '.join(row['cast'])
    director = ' '.join(row['director'])
    genres = ' '.join(row['genres'])
    return keywords + ' ' + cast + ' ' + director + ' ' + genres

df['input'] = df.apply(create_input, axis=1)
```

**CountVectorizer** converts a collection of text documents to a matrix of token counts, which can then be used for various machine learning tasks such as text classification, clustering, and recommendations. The stop_words parameter can be used to remove common words like "the", "and", "a", etc. from the input text.

```python
# Import CountVectorizer and create the count matrix
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df['input'])
```

**Cosine similarity** is a measure of similarity between two non-zero vectors of an inner product space that measures the cosine of the angle between them. In the context of text analysis, cosine similarity is commonly used to calculate the similarity between two documents based on the frequency of the words they share. A cosine similarity of 1 indicates that the two vectors are identical, while a cosine similarity of 0 indicates that the two vectors are orthogonal or completely dissimilar.

```python
# Compute the Cosine Similarity matrix based on the count_matrix

from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(count_matrix, count_matrix)
```

```python
df = df.reset_index()
indices = pd.Series(df.index, index=df['title'])
```

1. First, find the index of the movie in the dataframe that matches the given title.

2. Calculate the cosine similarity scores between this movie and all other movies in the dataframe. Store these scores as a list of tuples where the first element is the index of the other movie and the second element is the similarity score.
3. Sort the list of tuples based on the similarity scores in descending order.
4. Retrieve the top 10 tuples from the sorted list, excluding the first tuple as it corresponds to the movie itself.
5. Return the titles of the movies corresponding to the indices in the top 10 tuples.

```python
# Function that takes in movie title as input and outputs most similar movies
def get_recommendations(title, cosine_sim):
    # Get the index of the movie that matches the title
    idx = indices[title]
    
    # Get the pairwsie similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar movies
    sim_scores = sim_scores[1:11]

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar movies
    return df['title'].iloc[movie_indices]
```

```python
get_recommendations('Minions', cosine_sim)
```

```python
get_recommendations('Deadpool', cosine_sim)
```
